### 결측치 처리

```python
data.isna().sum()
```

**fuel_type 결측치 처리**

model 값이 동일한 행의 fuel_type 값을 넣어줌

```python
# 모델을 기준으로 정렬
data = data.sort_values('model')
data
```

```python
df = data.groupby('model', as_index = False)[['fuel_type']].count()
df
```

```python
# fuel_type 값이 NaN인데 동일한 model 값도 존재하지 않는 경우
df[df['fuel_type'] == 0]
```

```python
type(df)
```

```python
# fuel_type 열의 최빈값
data['fuel_type'].mode()
```

```python
for index, row in data.iterrows() :
    if (type(row.fuel_type) == str) and (len(row.fuel_type) >= 2) :
        continue
        
    else :  
        d = df.loc[df['model'] == row.model]
        
        if (d['fuel_type'].any() != 0) :
            try : 
                data.fuel_type.fillna(method = 'bfill', limit = 1, inplace = True)
            except :
                continue
                
        else :
            try :
                data.fuel_type.fillna('Gasoline', inplace = True)
            except :
                continue

data.isna().sum()
```

```python
data = data.sort_values('fuel_type')
data
```

```python
data['fuel_type'].unique()
```

```python
data = data.sort_values('model')
data
```

```python
# '–'로 잘못 표기된 NaN 값들을 NaN으로 바꿔주기
for index, row in data.iterrows() :
    if (type(row.fuel_type) == str) and (len(row.fuel_type) >= 2) :
        continue
        
    else :  
        data.loc[index, 'fuel_type'] = np.NaN
```

```python
data.isna().sum()
```

```python
# NaN 값들을 다시 한 번 bfill로 채워줌
for index, row in data.iterrows() :
    if (type(row.fuel_type) == str) and (len(row.fuel_type) >= 2) :
        continue
        
    else :
        try : 
            data.fuel_type.fillna(method = 'bfill', limit = 1, inplace = True)
        except :
            continue
```

```python
data = data.sort_values('fuel_type')
data
```

```python
data['fuel_type'].unique()
```

### 가변수화

```python
data.head()
```

**price 가변수화**

```python
# $ 지우기
for index, row in data.iterrows():
    data.loc[index, 'price'] = data.loc[index, 'price'].strip("$")
```

```python
# , 지우기
for index, row in data.iterrows():
    data.loc[index, 'price'] = data.loc[index, 'price'].replace(',', '')
```

```python
# str -> int 로 단위 바꾸기
data['price'] = data['price'].astype('int64')
# data.head()
```

```python
data.info()
```

**brand 가변수화**

```python
data['brand'].value_counts()
```

```python
data = data.sort_values('brand')

data.reset_index(drop = True, inplace = True)
# data.head()
```

```python
len(data)
```

```python
data['brand'].value_counts()
```

```python
k = 0
for i in range(0, len(data)) :
    print(i)
    if(i + 1 >= len(data)) :
        data.iloc[i, 0] = k
        break
    
    else :
        if(data.iloc[i, 0] == data.iloc[i + 1, 0]) :
#            print('case2')
#            print('k=', k)
            data.iloc[i, 0] = k
        
        else :
#            print('case3')
#            print('k=', k)
            data.iloc[i, 0] = k
            k = k + 1

# data.head()
```
